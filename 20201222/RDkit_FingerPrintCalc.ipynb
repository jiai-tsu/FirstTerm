```
TopologicalFingerprints = [FingerprintMols.FingerprintMol(df_ms[i]) for i in range(len(df_ms))]
```

```
from rdkit.Chem import MACCSkeys
MACCS_Keys = [MACCSkeys.GenMACCSKeys(df_ms[i]) for i in range(len(df_ms))]
```

```
from rdkit.Chem.AtomPairs import Pairs
AtomPairs_and_TopologicalTorsions = [Pairs.GetAtomPairFingerprint(df_ms[i]) for i in range(len(df_ms))]
```

```
from rdkit.Chem.AtomPairs import Torsions
TopologicalTorsion = [Torsions.GetTopologicalTorsionFingerprintAsIntVect(df_ms[i]) for i in range(len(df_ms))]
```

```
MorganFingerprint = [AllChem.GetMorganFingerprint(df_ms[i],2) for i in range(len(df_ms))]
```

```
MorganFingerprint2 = [AllChem.GetMorganFingerprintAsBitVect(df_ms[i],2,2048) for i in range(len(df_ms))]
MorganFingerprint2
```

```
df["Fsp3"] = df["ROMol"].map(Descriptors.FractionCSP3)
df['FP'] = df.apply(lambda x: AllChem.GetMorganFingerprintAsBitVect(x['ROMol'], 2, 2048), axis=1)
```

```
#フィンガープリントを再計算し、リスト作成
l_2d = df['ROMol'].values.tolist()
morgan_fp = [AllChem.GetMorganFingerprintAsBitVect(x,2,2048) for x in l_2d]
#類似度を計算し、距離行列？類似度の行列？を作成
dis_matrix = [DataStructs.BulkTanimotoSimilarity(morgan_fp[i], morgan_fp[:len(l_2d)],returnDistance=True) for i in range(len(l_2d))]
dis_array = np.array(dis_matrix)
print(dis_array)
```
